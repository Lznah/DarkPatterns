```python
import sqlite3
import pandas as pd

db = '/home/xhanpet/data/20210617-134559_segmentation_pilot/20210617-134559_segmentation_pilot.sqlite'
con = sqlite3.connect(db)
site_visits = pd.read_sql_query('''SELECT inner_text, site_url FROM (SELECT 
    visit_id, inner_text
FROM 
    segments WHERE inner_text LIKE "%notifikuj%") AS shop_segments JOIN site_visits USING(visit_id)''', con)
```

```python
site_visits['domain']=site_visits['site_url'].apply(lambda x: urlparse(str(x)).hostname)
site_visits['domain']=site_visits['domain'].apply(lambda x: x.replace('www.', ''))
```

```python
site_visits_unique_domain = site_visits.drop_duplicates(subset='domain')
```
